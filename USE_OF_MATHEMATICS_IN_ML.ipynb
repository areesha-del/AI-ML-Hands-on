**MEAN , MEDIAN,SD**

```python
import statistics

data = [10, 20, 30, 40, 50]

mean = statistics.mean(data)
median = statistics.median(data)
std_dev = statistics.stdev(data)   # sample standard deviation

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

**CORELATION**

```python
import numpy as np

x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

correlation = np.corrcoef(x, y)[0, 1]
print("Correlation:", correlation)
```

**CLASS WORK**

```python
import numpy as np

def summarize_clinic(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return mean, std_dev

# Given data
data = [4.2, 5.1, 3.8, 7.2, 5.5, 6.0]

mean, std = summarize_clinic(data)
print("Mean:", mean)
print("Standard Deviation:", std)
```

**CONTIIUE----**

```python
# Add outlier
data.append(15.0)

# Recalculate
new_mean, new_std = summarize_clinic(data)
print("New Mean:", new_mean)
print("New Standard Deviation:", new_std)
```

**DOSAGE VS RECOVERY WITH DIFFERENT WEIGHTS**

```python
import matplotlib.pyplot as plt
import numpy as np

dosage = np.array([10, 20, 30, 40, 50])
recovery = np.array([15, 25, 35, 45, 55])
plt.scatter(dosage, recovery, label="Actual Data")
w = 0.1
prediction = w * dosage
plt.plot(dosage, prediction, label="w = 0.1")
weights = [0.2, 0.5, 0.8]

for w in weights:
    plt.plot(dosage, w * dosage, label=f"w = {w}")
plt.xlabel("Dosage")
plt.ylabel("Recovery")
plt.title("Dosage vs Recovery with Different Weights")
plt.legend()
plt.show()

```

**TASK 5: MODEL LEARNING SIMULATION**

```python
# 1. Initialize weight and learning rate
w = 20
lr = 0.1

# 2. Reduce loss step by step
for i in range(10):
    loss = w**2              # Loss function
    gradient = 2 * w         # Derivative of w^2
    w = w - lr * gradient    # Update rule

    print(f"Step {i}, Weight {w:.2f}, Loss {loss:.2f}")
```
